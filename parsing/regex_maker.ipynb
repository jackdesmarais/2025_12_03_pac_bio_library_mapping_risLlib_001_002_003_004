```python
from Bio.Seq import Seq
```

```python
patterns_library = {
    "upstream_barcode":"CTGGCTAGCGNNNNNNGTGCACCTGA",
    "downstream_barcode": "GCTGCCTATCNNNNNNNNNNNNNNNNNNNNCTCGAGTCTA",
    "lib001_variable_exon":"CATCCAGGTT"+''.join(["N"]*100)+"AGGAGTAAGT",
    "lib001_variable_intron":"AGGAGTAAGT"+''.join(["N"]*40)+"GGTTAAGTTC",
    "lib002_variable_exon": "CATCCAGGTT"+''.join(["N"]*100)+"AGGAGTAAGC",
    "lib002_variable_intron":"AGGAGTAAGC"+''.join(["N"]*40)+"GGTTAAGTTC",
    "lib003_variable_exon":"TGGGCAGGTT"+''.join(["N"]*100)+"AGGACTCCTG",
    "lib004_junction":"CCCTGGGCAGCTCCTGGGCA"
    }
display(patterns_library)

patterns_fixed = {
    "CMV_enhancer":"CTTACGGTAA",
    "EX1":"gcaaggtgaa",
    "IN1":"gggcatatgg",
    "EX2_3ss":"tctctgccat",
    "EX2_SMN2_5ss":"aGGagtaagt",
    "EX2_ELP1_ris_5ss":"aGGagtaagc",
    "IN2":"atacataatc",
    'EX3':'tcactttggc',
    "polyA":"GCCACTCCCA",
    "ori":"TCCTGTTCCG",
    "ampR":"GGCTTACCAT",
    "ampR_promoter":"GAAGCATTTA",
    
}
display(patterns_fixed)
```

```python
import re
def repl(match):
    n_count = len(match.group(0))
    return f"(.{{1,{n_count*5}}})"

def pattern_maker(pattern_dict):
    new_patterns = {}
    for name, pattern in pattern_dict.items():
        pattern = pattern.upper()
        new_patterns[name] = re.sub(r'N+', repl, pattern)

        new_patterns[name+'_rev'] = re.sub(r'N+', repl, str(Seq(pattern).reverse_complement()))
    return new_patterns

new_lib_patterns = pattern_maker(patterns_library)
display(new_lib_patterns)
new_fixed_patterns = pattern_maker(patterns_fixed)
display(new_fixed_patterns)

```

```python
keys, values = zip(*new_lib_patterns.items())
print(str(keys).replace("'", '"').replace(", ", ' '))
print(str(values).replace("'", '"').replace(", ", ' '))
```

```python
keys, values = zip(*new_fixed_patterns.items())
print(str(keys).replace("'", '"').replace(", ", ' '))
print(str(values).replace("'", '"').replace(", ", ' '))
```
